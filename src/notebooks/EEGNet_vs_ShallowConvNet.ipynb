4. Definir el modelo
Puedes elegir entre EEGNet o ShallowConvNet desde models.py.

```python
# Sección 4: Definir modelo
model_eegnet = models.EEGNet()
model_shallow = models.ShallowConvNet()
```

5. Entrenamiento
Usa los scripts de referencia (train_eegnet_bci_iv2a.py, train_shallowconvnet_bci_iv2a.py).

En el notebook, define un loop de entrenamiento con PyTorch.

```python
# Sección 5: Entrenamiento
optimizer = torch.optim.Adam(model_eegnet.parameters(), lr=0.001)
loss_fn = torch.nn.CrossEntropyLoss()

for epoch in range(10):
    # forward
    outputs = model_eegnet(torch.tensor(X).float())
    loss = loss_fn(outputs, torch.tensor(y).long())
    
    # backward
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```
